```python
#Are Airport Trips More Expensive

df['near_airport'] = (
    (df['jfk_dist'] < 2) |
    (df['ewr_dist'] < 2) |
    (df['lga_dist'] < 2)
)

plt.figure()
sns.boxplot(x='near_airport',
            y='fare_amount',
            data=df)
plt.title("Fare for Airport vs Non-Airport Trips")
plt.show()
```

```python
#Are airport distance features redundant with total distance?

sample_df = df.sample(10000, random_state=42)

for col in ['jfk_dist', 'ewr_dist', 'lga_dist']:
    plt.figure()
    sns.scatterplot(x=sample_df['distance'],
                    y=sample_df[col],
                    alpha=0.3)
    plt.title(f"Distance vs {col}")
    plt.xlabel("Total Distance")
    plt.ylabel(col)
    plt.show()
```

```python
#Does trip distance vary across hours?

hourly_distance = df.groupby("hour")["distance"].mean()

plt.figure()
sns.lineplot(x=hourly_distance.index,
             y=hourly_distance.values)
plt.title("Average Distance by Hour")
plt.xlabel("Hour")
plt.ylabel("Mean Distance")
plt.show()
```

```python
#Is passenger count related to trip distance

plt.figure()
sns.boxplot(x='passenger_count',
            y='distance',
            data=df)
plt.title("Passenger Count vs Distance")
plt.show()
```

```python
avg_fare_passenger = df.groupby('passenger_count')['fare_amount'].mean()

plt.figure(figsize=(10,5))
sns.barplot(x=avg_fare_passenger.index,
            y=avg_fare_passenger.values)

plt.title("Average Fare by Passenger Count")
plt.xlabel("Passenger Count")
plt.ylabel("Average Fare")
plt.show()
```

```python
#Is calculated distance consistent with coordinate differences

df['lat_diff'] = abs(df['pickup_latitude'] - df['dropoff_latitude'])
df['lon_diff'] = abs(df['pickup_longitude'] - df['dropoff_longitude'])

plt.figure()
sns.scatterplot(x=df['lat_diff'],
                y=df['distance'],
                alpha=0.3)
plt.title("Latitude Difference vs Distance")
plt.show()
```
